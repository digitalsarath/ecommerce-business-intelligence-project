```python
sales_df = df[
    (
        ~df['Invoice']
        .astype(str)
        .str.startswith('C')
    )
    &
    (df['Quantity'] > 0)
    &
    (df['Price'] > 0)
].copy()

sales_df['Revenue'] = (
    sales_df['Quantity']
    *
    sales_df['Price']
)
```

### Create RFM Dataset

```python
analysis_date = pd.Timestamp(
    '2011-12-10'
)

rfm_source = sales_df[
    sales_df['Customer ID'] != 0
].copy()

rfm = rfm_source.groupby(
    'Customer ID'
).agg({
    'InvoiceDate': lambda x:
        (analysis_date - x.max()).days,
    'Invoice': 'nunique',
    'Revenue': 'sum'
}).reset_index()

rfm.columns = [
    'CustomerID',
    'Recency',
    'Frequency',
    'Monetary'
]
```

### RFM Scores

```python
rfm['R_Score'] = pd.qcut(
    rfm['Recency'],
    5,
    labels=[5,4,3,2,1]
)

rfm['F_Score'] = pd.qcut(
    rfm['Frequency']
    .rank(method='first'),
    5,
    labels=[1,2,3,4,5]
)

rfm['M_Score'] = pd.qcut(
    rfm['Monetary'],
    5,
    labels=[1,2,3,4,5]
)

rfm['RFM_Score'] = (
    rfm['R_Score'].astype(str)
    +
    rfm['F_Score'].astype(str)
    +
    rfm['M_Score'].astype(str)
)
```

### Customer Segmentation

```python
def segment_customer(row):

    if row['RFM_Score'] == '555':
        return 'Champions'

    elif int(row['R_Score']) >= 4 and int(row['F_Score']) >= 4:
        return 'Loyal Customers'

    elif int(row['R_Score']) >= 4:
        return 'Potential Loyalists'

    elif int(row['R_Score']) <= 2 and int(row['F_Score']) >= 4:
        return 'At Risk'

    else:
        return 'Others'

rfm['Segment'] = rfm.apply(
    segment_customer,
    axis=1
)
```

### Segment Summary

```python
segment_summary = (
    rfm['Segment']
    .value_counts()
)

print(segment_summary)
```

### Segment Revenue

```python
segment_revenue = (
    rfm.groupby('Segment')
    ['Monetary']
    .sum()
    .sort_values(
        ascending=False
    )
)

print(segment_revenue)
```

### Segment Percentage

```python
segment_percentage = (
    segment_revenue
    /
    segment_revenue.sum()
    * 100
)

print(
    segment_percentage.round(2)
)
```
